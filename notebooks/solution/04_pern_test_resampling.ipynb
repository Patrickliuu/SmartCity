```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import numpy as np
from scipy.interpolate import interp1d
from datetime import datetime, timedelta
```

```python
# set seed for reproducibility
np.random.seed(43)

# Set a start time for the time series
start_time = datetime(2023, 12, 9, 8, 0, 0)

# Number of data points
num_data_points = 20

# Generate random time intervals (some less than 5 minutes, others more)
# 5 minutes = 300 seconds
time_intervals = [np.random.randint(1, 600) for _ in range(num_data_points)]

# Generate timestamps
timestamps = [start_time]
for interval in time_intervals:
    timestamps.append(timestamps[-1] + timedelta(seconds=interval))

# Generate random values for each timestamp (for simplicity, let's use a range 0-100)
values = np.random.randint(0, 100, size=num_data_points + 1)

# Create a DataFrame
df = pd.DataFrame({'Timestamp': timestamps, 'Value': values})

# Set the Timestamp column as the index of the DataFrame (this is important for resampling)
df.set_index('Timestamp', inplace=True)
```

```python
# Resample using mean
resampled_mean = df.resample('5T').mean()

# Resample using meand and interpolation
resampled_mean_int = df.resample('5T').mean().interpolate(method='linear')

# Upsample to 1-minute intervals, apply linear interpolation, then downsample to 5-minute intervals
upsampled = df.resample('1T').mean()
interpolated = upsampled.interpolate(method='linear')#(method='spline', order=2)
resampled_up1intdown = interpolated.resample('5T').mean()

# Upsample to 2-minute intervals, apply linear interpolation, then downsample to 5-minute intervals
upsampled = df.resample('2T').mean()
interpolated = upsampled.interpolate(method='linear')#(method='spline', order=2)
resampled_up2intdown = interpolated.resample('5T').mean()

# Upsample to 3-minute intervals, apply linear interpolation, then downsample to 5-minute intervals
upsampled = df.resample('3T').mean()
interpolated = upsampled.interpolate(method='spline', order=2)# method='spline', order=2 or method='linear'
resampled_up3intdown = interpolated.resample('5T').mean()

# Upsample to 4-minute intervals, apply linear interpolation, then downsample to 5-minute intervals
upsampled = df.resample('4T').mean()
interpolated = upsampled.interpolate(method='linear')#(method='spline', order=2)
resampled_up4intdown = interpolated.resample('5T').mean()

# Upsample to 0.1-minute intervals, apply linear interpolation, then downsample to 5-minute intervals
upsampled = df.resample('0.1T').mean()
interpolated = upsampled.interpolate(method='linear')#(method='spline', order=2)
resampled_up01intdown = interpolated.resample('5T').mean()

# Resample using interpolation directly
resampled_interpolated = df.resample('5T').interpolate(method='linear') # method time also does not work


## Resample "manually" using scipy

# copy timestamps from 5-min resampling
resampled_scipy = resampled_mean.copy()

# interpolate with scipy
f = interp1d(df.index.astype(np.int64), df['Value'], kind='linear')
resampled_scipy['Value'] = f(resampled_scipy.index)


# Create a figure with subplots
fig, axs = plt.subplots(5, 1, figsize=(12, 8))
date_format = mdates.DateFormatter('%H:%M')
for ax in axs:
    ax.set_xticks(pd.date_range(df.index.min(), df.index.max(), freq='5T'))
    ax.grid(True, axis='x')
    ax.set_xticklabels([])
axs[-1].set_xticklabels(pd.date_range(df.index.min(), df.index.max(), freq='5T'), rotation=45)
axs[-1].xaxis.set_major_formatter(date_format)


# Mean
axs[0].plot(df.index, df['Value'], color='black', marker='x', linestyle='dotted', label='Original Series')
axs[0].plot(resampled_mean.index, resampled_mean['Value'], color='red', marker='.', label='Mean')
axs[0].set_title("df.resample('5T').mean()")
axs[0].legend()

# Mean+Interpolation
axs[1].plot(df.index, df['Value'], color='black', marker='x', linestyle='dotted', label='Original Series')
axs[1].plot(resampled_mean_int.index, resampled_mean_int['Value'], color='purple', marker='.', label='Mean+Interpolation')
axs[1].set_title("df.resample('5T').mean().interpolate(method='linear')")
axs[1].legend()

# Directly interpolated
axs[2].plot(df.index, df['Value'], color='black', marker='x', linestyle='dotted', label='Original Series')
axs[2].plot(resampled_interpolated.index, resampled_interpolated['Value'], color='blue', marker='.', label='Direct Linear Interpolation')
axs[2].set_title("df.resample('5T').interpolate(method='linear')")
axs[2].legend()

# Upsampled, interpolated, downsampled
axs[3].plot(df.index, df['Value'], color='black', marker='x', linestyle='dotted', label='Original Series')
axs[3].plot(resampled_up1intdown.index, resampled_up1intdown['Value'], color='green', marker='o', label='Upsampled 1 min')
# axs[3].plot(resampled_up2intdown.index, resampled_up2intdown['Value'], color='red', marker='.', label='Upsampled 2 min')
# axs[3].plot(resampled_up3intdown.index, resampled_up3intdown['Value'], color='blue', marker='.', label='Upsampled 3 min')
# axs[3].plot(resampled_up4intdown.index, resampled_up4intdown['Value'], color='magenta', marker='.', label='Upsampled 4 min')
# axs[3].plot(resampled_up01intdown.index, resampled_up1intdown['Value'], color='yellow', marker='.', label='Upsampled 0.1 min')
axs[3].set_title("df.resample('1T').mean(); upsampled.interpolate(method='linear'); interpolated.resample('5T').mean()")
axs[3].legend()

# scipy
axs[4].plot(df.index, df['Value'], color='black', marker='x', linestyle='dotted', label='Original Series')
axs[4].plot(resampled_scipy.index, resampled_scipy['Value'], color='magenta', marker='o', label='scipy')
axs[4].set_title("scipy.interp1d")
axs[4].legend()

plt.tight_layout()
plt.show()

# Results:
# resampling mean contains missing data
# mean+interpolation is complete, but does not nicely reproduce the data
# direct interploation yields completely wrong results
# upsampling, interpolation, downsampling is complete and looks okay (with spline  there was an outlier); need to upsample 2-3x faster than downsample
# scipy is complete and looks okay, but might not handle NaNs well
```
