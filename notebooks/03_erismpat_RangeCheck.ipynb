# Lets take care of the Metadata filtering

```python
with open('./notebooks/solution/temp/flow_rate_sensor_names.csv', 'r') as f:
    flow_rate_sensor_names = json.load(f)
```

```python
def metadata_filtering(StartDate: str = "2020-10-01", EndDate: str = "2020-10-31"):
    all_data = []  # Initialize an empty list to collect DataFrames
    for sensor in flow_rate_sensor_names:
        #print(sensor)
        df = GetMetaData(source_name=sensor, start_date=StartDate, end_date=EndDate)
        all_data.append(df)  # Collect each DataFrame
    all_data = pd.concat(all_data, ignore_index=True)  # Concatenate all DataFrames at once
    return all_data

df_meta = metadata_filtering()
```

# Lets use bokeh again

```python
from bokeh.plotting import figure, show, output_notebook, ColumnDataSource
from bokeh.models import HoverTool, Legend, LegendItem, BoxAnnotation
from bokeh.palettes import Category20
from bokeh.io import output_file

# Assuming df_october is your DataFrame with timestamps and sensor readings,
# and df_response is your DataFrame with the periods to disregard

# Convert the index of df_october to datetime if not already done
df_october.index = pd.to_datetime(df_october.index)

# Prepare the data for Bokeh
source = ColumnDataSource(df_october)

# Create a figure
p = figure(width=1200, height=500, x_axis_type="datetime",
           title="Flow rate Over Time", tools="xpan,xwheel_zoom,reset,save",
           background_fill_color="#efefef")

p.xaxis.axis_label = "Timestamp"
p.yaxis.axis_label = "Sensor Value"

# Add hover tool
hover = HoverTool(tooltips=[('Sensor', '$name'), ('Value', '@$name')])
p.add_tools(hover)

# Keep track of legend items
legend_items = []

# Plot each sensor data
colors = Category20[len(df_october.columns)]  # Adjusted for the number of columns in df_october
for i, column in enumerate(df_october.columns):
    line = p.line(x='timestamp', y=column, source=source, line_width=2, color=colors[i % len(colors)], name=column)
    legend_items.append(LegendItem(label=column, renderers=[line]))

# Highlight the disregard periods from df_response
for index, row in df_response.iterrows():
    if row['answer'] == 'disregard':
        box = BoxAnnotation(left=row['timestamp_start'], right=row['timestamp_end'],
                            fill_alpha=0.3, fill_color='red')
        p.add_layout(box)

# Create and add legend
legend = Legend(items=legend_items, click_policy="hide")
p.add_layout(legend, 'right')

# Output
# Uncomment the following line if you want to output to a static HTML file
# output_file("flow_rate.html")

# Display plots in the Jupyter Notebook
output_notebook()

# Display the plot
show(p)
```
